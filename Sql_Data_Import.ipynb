```python
# ============================================================
# Marketing Campaign Analysis
# Author: Swagata Barman
# Tool: Python (pandas, sqlalchemy) + MySQL
# ============================================================

# CELL 1 — Import Libraries
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# CELL 2 — Load CSV
df = pd.read_csv("marketing_campaign.csv", sep=";")
print(f"Dataset Shape: {df.shape}")
print(df.head())

# CELL 3 — Connect to MySQL
password = quote_plus("your_password")  # replace with your password
engine = create_engine(f"mysql+pymysql://root:{password}@127.0.0.1/marketing_analysis")

# Test connection
try:
    with engine.connect() as conn:
        print("✅ Connected to MySQL successfully!")
except Exception as e:
    print(f"❌ Connection failed: {e}")

# CELL 4 — Import Data into MySQL
df.to_sql("customers", con=engine, if_exists="replace", index=False)
print(f"✅ Done! {len(df)} rows imported into MySQL.")

# CELL 5 — Verify Import
df_check = pd.read_sql("SELECT * FROM customers LIMIT 5", con=engine)
print(df_check)
```
